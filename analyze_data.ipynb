```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

df = pd.read_csv("end_report.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
def plot_smt(name):
    curiosities = sorted(df["initial_curiosity"].unique())
    avg_knowledges = [df[df["initial_curiosity"]==c][name].mean() for c in curiosities]
    std_knowledges = [df[df["initial_curiosity"]==c][name].std() for c in curiosities]
    plt.errorbar(curiosities, avg_knowledges, yerr=std_knowledges, fmt='-o')
    plt.xlabel("Initial Curiosity")
    plt.ylabel(name)
    plt.title(name + " vs Initial Curiosity")
    plt.grid()
    plt.show()
plot_smt("avg_knowledge")
```

```python
plot_smt("explored_percentage")
```
